```python
PricebyYear= pd.pivot_table( price_2017, index=['시군구명','법정동명'],values=['공시지가(원/㎡)'] ) 
PricebyYear = PricebyYear.round(2)

PricebyYear=PricebyYear.rename(columns = {'공시지가(원/㎡)':'2017년'})

PricebyYear.head()
```

```python
PricebyYear.index.levels[0] 
```

```python
for year in range (2018,2021,1):
        
    path = ORIGIN_PATH + f'공시지가/공시지가_{year}년.csv' 
    target_file =pd.read_csv(path,encoding='cp949',sep=',')
    
    PricebyYear[f'{year}년']= pd.pivot_table( target_file, index=['시군구명','법정동명'],values=['공시지가(원/㎡)'] ) 
    PricebyYear = PricebyYear.round(2)
```

```python
top1 = PricebyYear.loc['관악구']
top1.T["남현동"]
```

```python

cols = top1.T.columns
cols
```

```python
plt.figure(figsize=(21,10))

plt.plot(top1.T)
# for col in cols:
#     plt.plot(top1.T[col])

plt.xlabel('측정 년도', fontsize = 20)
plt.ylabel('공시지가(원/㎡)',fontsize = 20)
    # 제목
plt.title('관악구의 공시지가(원/㎡)',fontsize = 20)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)

plt.legend(cols, loc='center left', bbox_to_anchor=(1, 0.5), fontsize = 20)
    
plt.savefig('result/price_top1.png')
plt.show()
```
