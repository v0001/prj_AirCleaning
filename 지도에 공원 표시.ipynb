```python
# [ lat, lng ] or [위도, 경도]
gps  = [35.163144,129.0633649]
# 지도를 띠울대 배율
zoom = 13
map  = folium.Map( location=gps, zoom_start=zoom )
map
```

```python

# 마커 생성
# 강남 파이낸스 센터 37.5003437,127.0361041
folium.Marker( gps, popup='역삼역 강남 파이낸스 센터', 
              icon=folium.Icon( icon='cloud' ) ).add_to( map )

folium.Marker( [37.5018564,127.0358954], popup='역삼역 KFC', 
              icon=folium.Icon( icon='info-sign', color='red' ) ).add_to( map )

# 서클의 크기는 배율에 따라 같이 커지고 작아진다
# 크기를 표현할때는 적절하나, 영역을 표현할때는 부적절하다
folium.CircleMarker( gps, popup='역삼역 SDS', radius=30, color='#3116cc', 
                    fill_color='#3116cc' ).add_to( map )              

map
```

```python
# 2017~2020 1월 데이터를 추출하여 지점명이 어떻게 변화하고 추가/삭제 되었는지 확인 
path = ORIGIN_PATH + '서울시 주요 공원현황.csv' 
target_df =pd.read_csv(path,encoding='cp949',sep=',')

target_df
```

```python

# 마커 생성
# 강남 파이낸스 센터 37.5003437,127.0361041

for idx in target_df.index:
    park_name = target_df['공원명'][idx]
    gps = [target_df['Y좌표(WGS84)'][idx],target_df['X좌표(WGS84)'][idx]]
    print(park_name,gps)
    folium.Marker( gps, popup=park_name, icon=folium.Icon( icon='info-sign', color='red')  ).add_to( map )
    
    
map
map.save('index.html')
```
